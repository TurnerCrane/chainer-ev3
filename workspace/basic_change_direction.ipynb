```python
import time
from lib.ev3 import EV3


# センサーとモーターの通信ポートを定義
touch_port = EV3.PORT_2
lmotor_port = EV3.PORT_B
rmotor_port = EV3.PORT_C

# センサーとモーターの設定
ev3 = EV3()
ev3.motor_config(lmotor_port, EV3.LARGE_MOTOR)
ev3.motor_config(rmotor_port, EV3.LARGE_MOTOR)
ev3.sensor_config(touch_port, EV3.TOUCH_SENSOR)

# タッチセンサーを押して離したことを確認したらスタート
print('Push touch sensor to run your EV3.')
while not ev3.touch_sensor_is_pressed(touch_port):
    pass
while ev3.touch_sensor_is_pressed(touch_port):
    pass

print('Go!')
ev3.motor_steer(lmotor_port, rmotor_port, 10, 0)
while True:
    # タッチセンサーを押すと１回目の方向変換
    if ev3.touch_sensor_is_pressed(touch_port):
        print('Changing the direction.')
        ev3.motor_steer(lmotor_port, rmotor_port, 10, 100)
        time.sleep(2)
        break

print('Go!')
ev3.motor_steer(lmotor_port, rmotor_port, 10, 0)
while True:
    # タッチセンサーを押すと2回目の方向変換
    if ev3.touch_sensor_is_pressed(touch_port):
        print('Changing the direction.')
        ev3.motor_steer(lmotor_port, rmotor_port, 10, 100)
        time.sleep(2)
        break

print('Go!')
ev3.motor_steer(lmotor_port, rmotor_port, 10, 0)
while True:
    # タッチセンサーを押すと3回目の方向変換
    if ev3.touch_sensor_is_pressed(touch_port):
        print('Changing the direction.')
        ev3.motor_steer(lmotor_port, rmotor_port, 10, 100)
        time.sleep(2)
        break

print('Go!')
ev3.motor_steer(lmotor_port, rmotor_port, 10, 0)
while True:
    # タッチセンサーを押すと4回目の方向変換
    if ev3.touch_sensor_is_pressed(touch_port):
        print('Changing the direction.')
        ev3.motor_steer(lmotor_port, rmotor_port, 10, 100)
        time.sleep(2)
        break

print('Go!')
ev3.motor_steer(lmotor_port, rmotor_port, 10, 0)
while True:
    # タッチセンサーを押すと停止
    if ev3.touch_sensor_is_pressed(touch_port):
        break

print("Stop.")
ev3.motor_steer(lmotor_port, rmotor_port, 0, 0)
ev3.close()
```
